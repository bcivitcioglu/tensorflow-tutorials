```python
def read_pkl(filename):
    path = filename.numpy().decode('utf-8')
    T = path.split('/')[-1].split('_')[8]
    with open(path, 'rb') as file:
        data = pkl.load(file)
        config = data['configuration']
        cor = data['correlation_length']
    return config, cor
```
